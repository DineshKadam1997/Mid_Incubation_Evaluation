```python
from PIL import Image, ImageDraw
import os
import random

# Function to generate a shape image with a random color
def generate_shape_image(width, height):
    image = Image.new("RGB", (width, height), "white")
    draw = ImageDraw.Draw(image)
    
    # Randomly choose a shape
    shape = random.choice(["rectangle", "ellipse", "polygon"])
    
    # Randomly choose a color
    color = random.choice(["red", "green", "blue"])
    if color == "red":
        fill_color = "red"
    elif color == "green":
        fill_color = "green"
    else:
        fill_color = "blue"
    
    # Draw the shape with the chosen color
    if shape == "rectangle":
        draw.rectangle([10, 10, width-10, height-10], fill=fill_color)
    elif shape == "ellipse":
        draw.ellipse([10, 10, width-10, height-10], fill=fill_color)
    elif shape == "polygon":
        points = [(random.randint(10, width-10), random.randint(10, height-10)) for _ in range(3)]
        draw.polygon(points, fill=fill_color)
    
    return image

# Function to save images to a directory
def save_images(num_images, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    for i in range(num_images):
        image = generate_shape_image(100, 100)
        image_path = os.path.join(output_dir, f"image_{i}.png")
        image.save(image_path)

# Generate and save 10 images to a directory named "shapes_dataset"
save_images(10, "/kaggle/working/shapes_dataset")
```

```python
from PIL import Image
import os

def convert_to_grayscale(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)  # Create the output directory if it doesn't exist
    for file_name in os.listdir(input_dir):
        if file_name.endswith('.png'):
            # Load the image
            image_path = os.path.join(input_dir, file_name)
            image = Image.open(image_path)
            
            # Convert to grayscale
            grayscale_image = image.convert('L')  # 'L' mode is for grayscale
            
            # Save the grayscale image
            output_path = os.path.join(output_dir, file_name)
            grayscale_image.save(output_path)

# Convert images in the 'shapes_dataset' directory to grayscale and save them to a new directory
convert_to_grayscale('/kaggle/working/shapes_dataset', '/kaggle/working/grayscale_dataset')
```

```python
import cv2
import os

def apply_edge_detection(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)  # Create the output directory if it doesn't exist
    for file_name in os.listdir(input_dir):
        if file_name.endswith('.png'):
            # Load the grayscale image
            image_path = os.path.join(input_dir, file_name)
            grayscale_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            
            # Apply edge detection
            edges = cv2.Canny(grayscale_image, 100, 200)  # Adjust the threshold values as needed
            
            # Save the edge-detected image
            output_path = os.path.join(output_dir, file_name)
            cv2.imwrite(output_path, edges)

# Apply edge detection to grayscale images in the 'grayscale_dataset' directory and save them to a new directory
apply_edge_detection('/kaggle/working/grayscale_dataset', '/kaggle/working/edge_detected_dataset')
```
